SHOWS THE PROPORTION OF SALES BY REGION 

```python
region_sales=df.groupby('Region')['Sales'].sum()
plt.figure(figsize=(5,5),facecolor='white')
plt.pie(region_sales,labels=region_sales.index,autopct='%1.1f%%',colors=plt.cm.Paired.colors)
plt.title('Sales by Region')
plt.show()
```

SHOWS THE TOTAL SALES FOR EACH PRODUCT CATEGORY

```python
category_sales = df.groupby('Product Category')['Sales'].sum()
bars = plt.bar(category_sales.index,category_sales)
for bar in bars:
    height=bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,height,f'{height:.0f}',ha='center',va='bottom')
plt.title('Sales by Product Cayegory')
plt.ylabel('Total Sales -->')
plt.xlabel('Product Catrgory -->')
plt.show()
```

DISPLAYS THE FREQUENCY DISTRIBUTION OF PROFIT

```python
plt.hist(df['Profit'],bins=10,color='green',edgecolor='black')
plt.title('Profit Distriibution')
plt.xlabel('Profit')
plt.ylabel('Distribution')
plt.show()
```

DISPLAYS THE RELATIONSHIP BETWEEN SHIPPING COST AND PROFIT

```python
plt.scatter(df['Shipping Cost'],df['Profit'],alpha=0.5,color='black')
plt.title('Shipping Cost vs Profit',color='blue')
plt.xlabel('Shipping Cost',color='red')
plt.ylabel('Profit',color='red')
plt.show()
```

SHOWS THE TREND OF SALES OVER TIME (MONTHLY)

```python
plt.figure(figsize=(8,4))
plt.plot(df["Order Date"],df["Sales"],marker='o')
plt.title('Sales Trend Over Time')
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,3))
df_sorted=df.sort_values(by="Order Date")
plt.plot(df_sorted["Order Date"],df_sorted["Sales"],marker='o',color='orange')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.scatter(df["Order Date"],df["Sales"],color='pink')
plt.grid(True)
plt.show()
```

VISUALIZES THE SPREAD AND DISTRIBUTION OF PROFIT BY PRODUCT CONTAINER

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x='Product Container',y='Profit')
plt.title('Profit by Product Container')
plt.xlabel('Product Container')
plt.ylabel('Profit')
plt.show()
```

DISPLAYS THE PROFIT AT PARTICULAR AMOUNT OF DISCOUNT APPLIED

```python
sns.scatterplot(data=df,x='Discount',y='Profit',color='black')
plt.title('Profit vs Discount')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```

SHOWS THE DISTRIBUTION OF PROFIT BY CUSTOMER SEGMENT

```python
sns.violinplot(data=df,x='Customer Segment',y='Profit',palette='muted')
plt.show()
```
